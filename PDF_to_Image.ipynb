```python
from pdf2image import convert_from_path
from glob import glob
```

```python
for f in glob('nms_files/*.pdf'):
    print('processing', f)
    images = (convert_from_path(f, fmt='png'))
    for i in range(len(images)):
        print('nms_files/images/'+ f.split('/')[1][:-4] + '_page_' + str(i+1) + '.png')
        images[i].save('nms_files/images/'+ f.split('/')[1][:-4]  + '_page_' + str(i+1) + '.png')
```
