```python
from google.cloud import vision
import io
import yaml
import os
from google.protobuf.json_format import MessageToJson
from glob import glob
import pandas as pd
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
with open('properties.yaml') as file:
    properties = yaml.full_load(file)

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = properties['google_application_credentials']
```

```python
client = vision.ImageAnnotatorClient()

page_text = []
page_confidence = []
file_name = []

for f in glob('nms_files/images/*.png'):
    with open(f, 'rb') as image_file:
        
        content = image_file.read()
        image = vision.Image(content=content)
        response = client.document_text_detection(image=image)
        
        text = response.full_text_annotation.text
        
        confidence = 0
        i = 0
        for p in response.full_text_annotation.pages:
            confidence += p.confidence
            i += 1
        #print(confidence/i)
        #print(text)
        
        #print()
        file_name.append(f)
        page_text.append(text)
        page_confidence.append(confidence/i)
        print(f, confidence)
        
        #break
    #break
        
```

```python
df = pd.DataFrame({'file':file_name,'text':page_text,'confidence':page_confidence})
```
